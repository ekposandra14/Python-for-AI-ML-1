```python
import pandas as pd
data = {'Month': [
        'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'
    ],
    'Laptops': [120, 135, 150, 160, 145, 170, 180, 175, 160, 155, 190, 200],
    'Smartphones': [200, 220, 250, 270, 260, 300, 310, 295, 280, 270, 320, 350],
    'Tablets': [90, 95, 100, 110, 105, 115, 120, 118, 112, 109, 125, 130],
    'Monitors': [60, 62, 65, 68, 66, 70, 75, 73, 72, 70, 78, 80],
    'Keyboards': [40, 42, 45, 47, 46, 48, 50, 49, 47, 46, 51, 55],
    'Mouse': [38, 39, 41, 42, 41, 43, 44, 43, 42, 41, 45, 47]
}
df = pd.DataFrame(data)
print(df.head())
```

```python
df = pd.DataFrame(data)
print(df.head())
import matplotlib.pyplot as plt
# sum  yearly totals, skipping the 'month' column
yearly_totals = df.sum(numeric_only=True)
# plot as bar
yearly_totals.plot(kind='bar',color='skyblue')
plt.xlabel('products')
plt.ylabel('Total sales (year)')
plt.title('Total Yearly sales per product')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df.plot (x='Month', y='Smartphones', kind='line', marker='o',color='skyblue')
plt.xlabel('Month')
plt.ylabel('Smartphone sale')
plt.title('monthly Smartphones Sales trends')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df['Tablets'].plot(kind='hist',bins=5, color='skyblue')
plt.xlabel('Tablets sale')
plt.ylabel('Frequency')
plt.title('Distribution of Tablets Sales')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df.plot(x='Mouse', y='Keyboards', kind='scatter', color='green')

plt.xlabel('Mouse sale')
plt.ylabel('Keyboards sale')
plt.title('Mouse vs Keyboards Sales')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df.to_csv('tech_store_sales.csv', index=False)

```
